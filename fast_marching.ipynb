```python
# Python3 implementation of Min Heap

import sys
from itertools import repeat

class MinHeap:

    def __init__(self, maxsize):
      self.maxsize = maxsize
      self.size = 0
      self.Heap = [[]for i in repeat(None,maxsize)]
      for i in range (maxsize):
        self.Heap[i].append(i)
        self.Heap[i].append(float(sys.maxsize))
      self.Heap.insert(0,[0,(-sys.maxsize)])
      self.FRONT = 1
      self.Heap = np.array(self.Heap)

    # Function to return the position of
    # parent for the node currently
    # at pos
    def parent(self, pos):
        return pos//2

    # Function to return the position of
    # the left child for the node currently
    # at pos
    def leftChild(self, pos):
        return 2 * pos

    # Function to return the position of
    # the right child for the node currently
    # at pos
    def rightChild(self, pos):
        return (2 * pos) + 1

    # Function that returns true if the passed
    # node is a leaf node
    def isLeaf(self, pos):
        return pos*2 > self.size

    # Function to swap two nodes of the heap
    def swap(self, fpos, spos):
      
      self.Heap[fpos],self.Heap[spos]  = self.Heap[spos],self.Heap[fpos]
      
      

    # Function to heapify the node at pos
    def minHeapify(self, index):

        # If the node is a non-leaf node and greater
        # than any of its child

        if not self.isLeaf(index):
            if (self.Heap[index][1] > self.Heap[self.leftChild(index)][1] or
            self.Heap[index][1] > self.Heap[self.rightChild(index)][1]):

                # Swap with the left child and heapify
                # the left child
                if self.Heap[self.leftChild(index)][1] < self.Heap[self.rightChild(index)][1]:
                    self.swap(index, self.leftChild(index))
                    self.minHeapify(self.leftChild(index))

                # Swap with the right child and heapify
                # the right child
                else:
                    self.swap(index, self.rightChild(index))
                    self.minHeapify(self.rightChild(index))

    #Function to insert a node into the heap
    def remove_node(self,node):
      index = np.where(self.Heap==node)[0][0]
      self.Heap[index][1]=-1*sys.maxsize
      while self.Heap[index][1] < self.Heap[self.parent(index)][1]:
          self.swap(index, self.parent(index))
          index = self.parent(index)
      self.remove()



    def insert(self, element):
        index = np.where(self.Heap==element[0])[0][0]
        self.Heap[index][1]=-1*sys.maxsize
        while self.Heap[index][1] < self.Heap[self.parent(index)][1]:
            self.swap(index, self.parent(index))
            index = self.parent(index)
        self.remove()


        if self.size >= self.maxsize+10 :
            return
        self.size+= 1
        self.Heap[self.size] = element #element is an array with index and and priority

        current = self.size
        print('current is',current )

        print('self.Heap[current] is',self.Heap[current] )

        # print('self.Heap[self.parent(current)] is',self.Heap[self.parent(current)] )

        while self.Heap[current][1] < self.Heap[self.parent(current)][1]:
            self.swap(current, self.parent(current))
            current = self.parent(current)
        
        


    #if it is the existing node it inserts only the priority
    def insert_element(self,node,value):

      #in this position the Heap has this ""node"
      index = np.where(self.Heap==node)[0][0]
      print('index where the node is present in the Heap: ',index)


      if self.Heap[index][0]==node:
        self.Heap[index][1]=value
        print('the value of Heap at the index is : ',self.Heap[index])
        print('the value at parent of the index is : ',self.Heap[self.parent(index)][1])
        #self.minHeapify(index)
        print(self.Heap[index][1]<self.Heap[self.parent(index)][1])

        while self.Heap[index][1]<self.Heap[self.parent(index)][1]:

          self.Heap[index],self.Heap[self.parent(index)]  = self.Heap[self.parent(index)],self.Heap[index]
          # self.swap(index,self.parent(index))
          index= self.parent(index)
          
      else:
        print("enter the node that already exists")


    # Function to print the contents of the heap
    def Print(self):
        for i in range(1, (self.size//2)+1):
            print(" PARENT : "+ str(self.Heap[i])+" LEFT CHILD : "+
                                str(self.Heap[2 * i])+" RIGHT CHILD : "+
                                str(self.Heap[2 * i + 1]))

    # Function to build the min heap using
    # the minHeapify function
    def minHeap(self):

        for pos in range(self.size//2, 0, -1):
            self.minHeapify(pos)

    # Function to remove and return the minimum
    # element from the heap
    def remove(self):

        popped = self.Heap[self.FRONT]
        self.Heap[self.FRONT] = self.Heap[self.size]
        self.size-= 1
        self.minHeapify(self.FRONT)
        return popped

# # Driver Code
# if __name__ == "__main__":
    
#     print('The minHeap is ')
#     minHeap = MinHeap(15)
#     minHeap.insert([1,5])
#     minHeap.insert([2,3])
#     minHeap.insert([3,17])
#     minHeap.insert([4,10])
#     minHeap.insert([5,84])
#     minHeap.insert([6,19])
#     minHeap.insert([7,6])
#     minHeap.insert([8,22])
#     minHeap.insert([9,9])
#     print(minHeap.Heap)
```

# **to find if there are obtuse angles**

```python
import math
from math import pi

def find_angle(A,b,c):
   #finds angle around the point A
   cosine_angle = np.dot(np.array(b)-np.array(A), np.array(c)-np.array(A)) / (np.linalg.norm(np.array(b)-np.array(A)) * np.linalg.norm(np.array(c)-np.array(A)))
   angle = np.arccos(cosine_angle)
   return angle



def find_obtuse(tri):
  i=0;
  
  obtuse_array = []
  for j in range (len(tri)):
    A = find_angle(vert[tri[j][0]],vert[tri[j][2]],vert[tri[j][1]])
    B = find_angle(vert[tri[j][1]],vert[tri[j][2]],vert[tri[j][0]])
    C = find_angle(vert[tri[j][2]],vert[tri[j][0]],vert[tri[j][1]])

    obtuse= False

    if A>(pi/2):
      obtuse= True
      
    if B>(pi/2):
      obtuse= True
      
    if C>(pi/2):
      obtuse= True

    if obtuse == True:
      obtuse_array.append(j)
      
  obtuse_array =np.array(obtuse_array)
  return obtuse_array 

print(len(find_obtuse(tri)))

        
```

# **updater**

```python
from math import sqrt
def calc_distance(p1,p2):
  distance = sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2+(p1[2]-p2[2])**2)
  return distance

# d1= calc_distance(vert[0],vert[1])
# d2= calc_distance(vert[0],vert[2])
#finds the update procedure when there are 2 points given and 
#d1 is distance to x1 and d2 to x2 and d3 to x3 from  a source point
#x1 would be a vert[1]for example has to have a xyz component
def updater(x1,d1,x2,d2,x3,d3):
  u = abs(d1-d2)
  if d1>d2:
    a = calc_distance(x1,x3)
    b = calc_distance(x2,x3)
    
  if d2>=d1:
    b = calc_distance(x1,x3)
    a = calc_distance(x2,x3)
  
  theta = find_angle(x3,x2,x1)
  #print('variable a is ', a)
  A= a**2+b**2-2*a*b*math.cos(theta)
  B= 2*b*u*(a*math.cos(theta)-b)
  C= b**2*(u**2-a**2*math.sin(theta)**2)

  root_term = B**2-4*A*C

  if root_term < 0 :
    root_term =0



  t1= (-B+sqrt(root_term))/(2*A)
  t2= (-B-sqrt(root_term))/(2*A)

  t= max(t1,t2)

  if (u<t and a*math.cos(theta)<(b*(t-u)/t) and (b*(t-u)/t)< a/math.cos(theta)):
    d3 = min(d3,t+b)
  else:
    d3=min(d3,b+min(d1,d2),a+max(d1,d2))

  return d3

d1= 0.007333421810442922
d2= 0.0024707027817164866
print(updater(vert[46353],d1,vert[46354],d2,vert[46352],1000))
print (calc_distance(vert[0],vert[46355]))


# x1 is  46353
# d1 is  0.007333421810442922
# x2 is  46354
# d2 is  0.0024707027817164866
# x3 is  46352
```
